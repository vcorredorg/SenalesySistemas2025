###Listas

```python
#Escribe una función que reciba una lista de números como entrada y devuelva la suma de todos los números impares de la lista.
def sumaImpares(lista):
  suma=0
  for x in range(len(lista)):
    if lista[x]%2!=0:
      suma+=lista[x]
  return suma

lista=[1,2,3,4,5]
print(sumaImpares(lista))
```

```python
#Escribe una función que reciba una lista de palabras como entrada y devuelva un diccionario que cuente la cantidad de veces que aparece cada palabra en la lista
def cuentaPalabras(lista):
  diccionario={}
  for x in range(len(lista)):
    if lista[x] in diccionario:
      diccionario[lista[x]]+=1
    else:
      diccionario[lista[x]]=1
  return diccionario
```

```python
#Crea una clase "Rectángulo" que tenga atributos como longitud y ancho y métodos para calcular el área y el perímetro
class Rectangulo:
  def __init__(self, longitud, ancho):
    self.longitud= longitud
    self.ancho= ancho

  def area(self, longitud, ancho):
    area= self.longitud* self.ancho
    return area

  def perimetro(self, longitud, ancho):
    perim= 2*self.longitud + 2*ancho
    return perim

rectangulo= Rectangulo(2, 4)
print(rectangulo.area(rectangulo.longitud, rectangulo.ancho))
rectangulo.perimetro(rectangulo.longitud, rectangulo.ancho)
```

```python
#Crea una clase "Banco" que tenga atributos como nombre, dirección y métodos para abrir una cuenta, depositar dinero, retirar dinero y revisar el saldo
class Banco:
  def __init__(self, nombre, direccion, saldo):
    self.nombre= nombre
    self.direccion= direccion
    self.saldo= saldo

  def abrirCuenta(self, saldo):
    print(f"Cuenta abierta con saldo de {self.saldo}")

  def depositar(self, monto):
    if monto> self.saldo:
      print("Saldo insuficiente")
    else:
      restante= self.saldo-monto
      print(f"Saldo restante: {restante}")

  def retirar(self, monto):
    if monto> self.saldo:
      print("Saldo insuficiente")
    else:
      restante= self.saldo-monto
      print(f"Saldo restante: {restante}")

  def revisarSaldo(self):
    print(f"Saldo actual: {self.saldo}")
```
