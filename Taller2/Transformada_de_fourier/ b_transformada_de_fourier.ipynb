###**1.3Encuentre la función de densidad espectral (Transformada de Fourier) para las siguientes señales, sin aplicar propiedades**:

a) $$e^{-a|t|},\quad a \in \mathbb{R}^+$$

b) $$\cos(\omega_c t),\quad \omega_c \in \mathbb{R}$$

c) $$\sin(\omega_s t),\quad \omega_s \in \mathbb{R}$$

d) $$f(t)\cos(\omega_c t),\quad \omega_c \in \mathbb{R},\ f(t)\in \mathbb{R}, \text{ continua}$$

e) $$e^{-a|t|^2},\quad a \in \mathbb{R}^+$$

f) $$A\,\mathrm{rect}\!\left(\frac{t}{d}\right),\quad A,d \in \mathbb{R}$$

Convención y recordatorios

Usaremos la convención (frecuencia angular ω):
$$
X(\omega) = \mathcal{F}\{x(t)\} = \int_{-\infty}^{\infty} x(t)\, e^{-j\omega t}\, dt
$$

$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\omega)\, e^{j\omega t}\, d\omega
$$

(a) $$x(t)=e^{-a|t|},\ a>0$$

Partimos de la definición y partimos el eje temporal en dos:

$$
X(\omega)=\int_{-\infty}^{0} e^{at} e^{-j\omega t} dt \;+\; \int_{0}^{\infty} e^{-at} e^{-j\omega t} dt.
$$

Evaluando cada integral (geometría exponencial):

$$
\int_{0}^{\infty} e^{-(a+j\omega)t} dt = \frac{1}{a + j\omega}, \qquad
\int_{-\infty}^{0} e^{(a-j\omega)t} dt = \frac{1}{a - j\omega}.
$$

Sumando:

$$
X(\omega)=\frac{1}{a+j\omega}+\frac{1}{a-j\omega}
=\frac{(a-j\omega)+(a+j\omega)}{a^2+\omega^2}
=\frac{2a}{a^2+\omega^2}.
$$


(b) $$x(t)=\cos(\omega_c t)$$

Definición truncada y límite:

$$
X(\omega)=\lim_{L\to\infty}\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt.
$$

Escribimos $$e^{-j\omega t}=\cos(\omega t)-j\sin(\omega t)$$ y separamos:

$$
\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt
=\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt
 - j\int_{-L}^{L}\cos(\omega_c t)\sin(\omega t)\, dt.
$$

$$\cos(\cdot)\sin(\cdot)$$ es impar en \([ -L, L ] \Rightarrow\) su integral es \(0\).

Queda la parte real:

$$
\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt.
$$

Usamos la identidad trigonométrica (no es propiedad de FT):

$$
\cos A \cos B = \tfrac{1}{2}\big[\cos(A-B)+\cos(A+B)\big].
$$

Entonces

$$
\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt
=\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_c-\omega)t\big)\, dt
+\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_c+\omega)t\big)\, dt.
$$

Como $$\int_{-L}^{L}\cos(k t)\, dt = \frac{2\sin(kL)}{k},$$ obtenemos

$$
\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt
=\frac{\sin\big((\omega_c-\omega)L\big)}{\omega_c-\omega}
+\frac{\sin\big((\omega_c+\omega)L\big)}{\omega_c+\omega}.
$$

Paso al límite (sentido de distribuciones):

$$
\lim_{L\to\infty}\frac{\sin(L x)}{\pi x} = \delta(x).
$$

Por tanto,

$$
X(\omega)=\pi\big[\delta(\omega-\omega_c)+\delta(\omega+\omega_c)\big].
$$




(c) $$x(t)=\sin(\omega_s t)$$

Definición truncada y límite:

$$
X(\omega)=\lim_{L\to\infty}\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt.
$$

Separando:

$$
\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt
=\int_{-L}^{L}\sin(\omega_s t)\cos(\omega t)\, dt
- j\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt.
$$

$$\sin(\cdot)\cos(\cdot)$$ es impar \(\Rightarrow\) integral \(=0\).

Queda la parte imaginaria:

$$
- j\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt.
$$

Con la identidad

$$
\sin A \sin B = \tfrac{1}{2}\big[\cos(A-B)-\cos(A+B)\big],
$$

tenemos

$$
\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt
=\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_s-\omega)t\big)\, dt
-\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_s+\omega)t\big)\, dt.
$$

Usando \(\displaystyle\int_{-L}^{L}\cos(k t)\, dt = \frac{2\sin(kL)}{k}\):

$$
\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt
=-j\Big[\frac{\sin\big((\omega_s-\omega)L\big)}{\omega_s-\omega}
-\frac{\sin\big((\omega_s+\omega)L\big)}{\omega_s+\omega}\Big].
$$

Paso al límite (distribuciones):

$$
\lim_{L\to\infty}\frac{\sin(L x)}{\pi x} = \delta(x)
\;\Longrightarrow\;

$$
X(\omega) = \frac{\pi}{j}\,\big[\,\delta(\omega-\omega_s)\;-\;\delta(\omega+\omega_s)\,\big].
$$





(d) $$x(t)=f(t)\cos(\omega_c t),\quad f(t)\in\mathbb{R}\ \text{o}\ \mathbb{C}$$

Definimos  
$$F(\Omega)=\int_{-\infty}^{\infty} f(t)\,e^{-j\Omega t}\,dt.$$

Desde la definición:

$$
X(\omega)=\int_{-\infty}^{\infty} f(t)\cos(\omega_c t)\,e^{-j\omega t}\,dt
$$

$$
=\frac{1}{2}\int f(t)\,e^{-j(\omega-\omega_c)t}\,dt
+\frac{1}{2}\int f(t)\,e^{-j(\omega+\omega_c)t}\,dt
$$

$$
=\frac{1}{2}\,F(\omega-\omega_c)+\frac{1}{2}\,F(\omega+\omega_c).
$$

Esto se obtuvo directamente de la definición (simplemente reagrupando el exponente).


(e) $$x(t)=e^{-a\,|t|^{2}}=e^{-a t^{2}},\quad a>0$$

$$
X(\omega)=\int_{-\infty}^{\infty} e^{-a t^{2}}\, e^{-j\omega t}\, dt
= \int e^{-a\left[t^{2}+\frac{j\omega}{a}t\right]}\, dt
$$

$$
= \int \exp\!\left(-a\left(t+\frac{j\omega}{2a}\right)^{2}\right)\,
\exp\!\left(-\frac{\omega^{2}}{4a}\right)\, dt
$$

$$
= \exp\!\left(-\frac{\omega^{2}}{4a}\right)\,
\int_{-\infty}^{\infty} \exp\!\left(-a\left(t+\frac{j\omega}{2a}\right)^{2}\right)\, dt
$$

La integral gaussiana (desplazada) vale $\pi/a$. Por tanto,

$$
X(\omega)=\sqrt{\frac{\pi}{a}}\,\exp\!\left(-\frac{\omega^{2}}{4a}\right).
$$


(f) $$x(t)=A\,\mathrm{rect}_d(t)$$ (pulso de ancho \(d>0\) y amplitud \(A\))

Por definición:

$$
x(t)=
\begin{cases}
A, & |t|<\dfrac{d}{2},\\[6pt]
0, & \text{en otro caso}.
\end{cases}
$$

Entonces

$$
X(\omega)=\int_{-d/2}^{d/2} A\,e^{-j\omega t}\,dt
= A\left[\frac{e^{-j\omega t}}{-j\omega}\right]_{-d/2}^{d/2}
$$

$$
= A\;\frac{e^{-j\omega d/2}-e^{j\omega d/2}}{-j\omega}
= A\;\frac{-2j\sin(\omega d/2)}{-j\omega}
$$

$$
= A\;\frac{2\sin(\omega d/2)}{\omega}
= A\,d\;\mathrm{sinc}\!\left(\frac{\omega d}{2}\right).
$$

En ω=0, por continuidad, $$X(0)=A\,d.$$

###**Propiedades útiles de la Transformada de Fourier**

$$
\text{(1) Modulación:} \quad \mathcal{F}\{e^{-j\omega_1 t} \, x(t)\} = X(\omega + \omega_1)
$$

$$
\text{(2) Desplazamiento en Frecuencia (Inversa):} \quad
\mathcal{F}^{-1}\{X(\omega - \omega_0)\} = e^{j\omega_0 t} x(t)
$$

$$
\text{(3) Producto por } t^n: \quad
\mathcal{F}\{t^n x(t)\} = j^n \frac{d^n}{d\omega^n} X(\omega)
$$

$$
\text{(4) Paso Unitario:} \quad
\mathcal{F}\{u(t)\} = \text{PV}\!\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)
$$

$$
\text{(5) Exponencial Bilateral:} \quad
\mathcal{F}\{e^{-a|t|}\} = \frac{2a}{a^2 + \omega^2}, \quad a>0
$$

$$
\text{(6) Exponencial Unilateral:} \quad
\mathcal{F}\{u(t)e^{-at}\} = \frac{1}{a + j\omega}, \quad a>0
$$

$$
\text{(7) Coseno:} \quad
\mathcal{F}\{\cos(\omega_c t)\} = \pi\left[\delta(\omega - \omega_c) + \delta(\omega + \omega_c)\right]
$$

$$
\text{(8) Periodización en Frecuencia:} \quad
\mathcal{F}^{-1}\left\{\sum_{n=-\infty}^{\infty} X(\omega - n\omega_0)\right\}
=
T \sum_{k=-\infty}^{\infty} x(kT)\,\delta(t - kT), \quad \omega_0 = \frac{2\pi}{T}
$$



**(a)**

$$
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
$$

Usamos la identidad:

$$
\cos^2(\theta)=\frac{1}{2}\left(1+\cos(2\theta)\right)
$$

Entonces:

$$
u(t)\cos^2(\omega_c t)
=
\frac{1}{2}u(t)
+
\frac{1}{2}u(t)\cos(2\omega_c t)
$$

Aplicando transformada y la propiedad de modulación:

$$
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
=
\frac{1}{2}\mathcal{F}\{u(t)\}
+
\frac{1}{2}\cdot\frac{1}{2}\left[U(\omega - 2\omega_c)+U(\omega + 2\omega_c)\right]
$$

donde:

$$
U(\omega)=\mathcal{F}\{u(t)\}
=
\text{PV}\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)
$$

Así, el resultado final es:

$$
\boxed{
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
=
\frac{1}{2}\left[\text{PV}\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)\right]
+
\frac{1}{4}\sum_{s=\pm1}
\left[
\text{PV}\left\{\frac{1}{j(\omega + s\,2\omega_c)}\right\}
+
\pi\delta(\omega + s\,2\omega_c)
\right]
}
$$


**(b)**

$$\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}$$

Por modulación directa:

$$
\mathcal{F}\{\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - \omega_c)+\delta(\omega + \omega_c)\right].
$$

Entonces, usando la propiedad:

$$
\mathcal{F}\{e^{-j\omega_1 t}x(t)\}=X(\omega+\omega_1),
$$

se obtiene:

$$
\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - (\omega_c - \omega_1)) + \delta(\omega + (\omega_c + \omega_1))\right].
$$

$$
\boxed{\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - (\omega_c - \omega_1)) + \delta(\omega + (\omega_c + \omega_1))\right]}
$$


**(c)**

$$\mathcal{F}\{3t^3\}$$

Sabemos que:

$$x(t)=1 \quad \Rightarrow \quad X(\omega)=2\pi\delta(\omega).$$

Usando la propiedad del producto por potencias de \(t\):

$$
\mathcal{F}\{t^n x(t)\} = j^n \frac{d^n}{d\omega^n}X(\omega),
$$

entonces, para \(t^3\):

$$
\mathcal{F}\{t^3\}
=
j^3 \frac{d^3}{d\omega^3}\{2\pi \delta(\omega)\}
=
-\,j \, 2\pi\, \delta^{(3)}(\omega).
$$

Multiplicando por 3:

$$
\mathcal{F}\{3t^3\}
=
3\cdot\left(-\,j \, 2\pi\, \delta^{(3)}(\omega)\right)
=
-\,6\pi j \,\delta^{(3)}(\omega).
$$

$$
\boxed{\mathcal{F}\{3t^3\} = -6\pi j\,\delta^{(3)}(\omega)}
$$


**(d)**

$$\mathcal{F}^{-1}\left\{\frac{7}{\omega^2 + 6\omega + 45}\right\}$$

Completamos cuadrado:

$$
\omega^2 + 6\omega + 45 = (\omega+3)^2 + 6^2.
$$

Usamos la propiedad:
$$
\text{Si } X(\omega)=\frac{1}{a^{2}+(\omega-\omega_{0})^{2}}, \quad a>0,
$$

$$
\text{entonces la transformada inversa } x(t)=\frac{1}{2a}\,e^{-a|t|}\,e^{j\omega_{0}t}.
$$


Aquí:

$$a = 6, \qquad \omega_0 = -3.$$

Entonces:

$$
\mathcal{F}^{-1}\left\{\frac{7}{(\omega+3)^2 + 6^2}\right\}
=
7 \cdot \frac{1}{2\cdot 6} \, e^{-6|t|} \, e^{-j3t}
=
\frac{7}{12} e^{-6|t|} e^{-j3t}.
$$

$$
\boxed{\mathcal{F}^{-1}\left\{\frac{7}{\omega^2 + 6\omega + 45}\right\} = \frac{7}{12}\,e^{-6|t|}\,e^{-j3t}}
$$


### (e)

Dado:
$$
F^{-1}\left\{\frac{B}{T}\sum_{n=-\infty}^{\infty}
\left[
\frac{1}{a^{2}+(\omega-n\omega_{0})^{2}}

F^{-1}\left\{\sum_{n=-\infty}^{\infty}X_{2}(\omega-n\omega_{0})\right\}

$$

$$
\Rightarrow\quad
\boxed{

**1.5 Consulte en qué consiste la modulación por amplitud por
detección coherente y sus aplicaciones. Genere un ejemplo
ilustrativo sobre Python en el que se grafique las señales en
el tiempo y en frecuencia (utilizando la ‘rfft‘) para: señal
mensaje tipo pulso rectangular y señal mensaje tipo coseno.
El usuario podrá definir el índice de modulación de interés.
Ver cuaderno Modulación AM**

¿Qué es AM con detección coherente?

Señal AM (portadora no suprimida):

$$
s(t)=A_c\,[1+\mu\,m(t)]\cos(2\pi f_c t),\quad 0\le \mu \le 1.
$$

Modulación por amplitud (AM): la amplitud de la portadora
$$
\cos(2\pi f_c t)
$$
varía según el mensaje
$$
m(t).
$$

Detección coherente (sincrónica): en el receptor se multiplica por un oscilador local en fase y a la misma frecuencia que la portadora:

$$
y(t)=2\,s(t)\cos(2\pi f_c t)\ \xrightarrow{\text{LPF}}\ A_c+\mu A_c\,m(t).
$$

Tras el filtro pasa-bajos (LPF) se extrae el término de baja frecuencia (el mensaje, más una componente DC \(A_c\)).  
Si se desea el mensaje puro, se elimina la componente DC.

Aplicaciones: radiodifusión AM, telemetría, comunicaciones aeronáuticas heredadas, y sistemas donde la simplicidad del transmisor/receptor es importante.  
La detección coherente es necesaria para DSB-SC (portadora suprimida) y mejora la robustez frente a distorsiones de envolvente.


```python
import numpy as np
import matplotlib.pyplot as plt

# Intentamos usar scipy para un LPF cómodo; si no está, creamos uno FIR sencillo
try:
    from scipy.signal import butter, filtfilt
    HAVE_SCIPY = True
except Exception:
    HAVE_SCIPY = False

def rfft_spectrum(x, fs):
    """
    Devuelve f (Hz) y |X| escalado de una rfft monolado.
    """
    N = len(x)
    X = np.fft.rfft(x)
    f = np.fft.rfftfreq(N, d=1/fs)
    # Escala monolado (amplitud): 2/N excepto DC y Nyquist (si existe)
    mag = np.abs(X) / N
    if N % 2 == 0:
        mag[1:-1] *= 2
    else:
        mag[1:] *= 2
    return f, mag

def simple_lpf(x, fs, fc_hz):
    """
    LPF. Usa Butterworth+filtfilt si SciPy está disponible; si no, FIR por ventana-sinc.
    """
    if fc_hz <= 0:
        return np.zeros_like(x)
    if HAVE_SCIPY:
        b, a = butter(4, fc_hz/(fs/2.0), btype='low')
        return filtfilt(b, a, x)
    # FIR ventana-sinc (fallback)
    num_taps = 401
    n = np.arange(num_taps) - (num_taps-1)/2
    h = np.sinc(2*fc_hz*n/fs)  # ideal LPF
    w = np.hamming(num_taps)
    h = h * w
    h = h / np.sum(h)
    return np.convolve(x, h, mode='same')

def plot_time(t, signals, labels, xlim=None, title=''):
    plt.figure(figsize=(10,3))
    for s, lab in zip(signals, labels):
        plt.plot(t, s, label=lab)
    if xlim is not None:
        plt.xlim(xlim)
    plt.xlabel('Tiempo [s]')
    plt.ylabel('Amplitud')
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

def plot_freq(f, mags, labels, xlim=None, title=''):
    plt.figure(figsize=(10,3))
    for M, lab in zip(mags, labels):
        plt.plot(f, M, label=lab)
    if xlim is not None:
        plt.xlim(xlim)
    plt.xlabel('Frecuencia [Hz]')
    plt.ylabel('Amplitud (monolado)')
    plt.title(title)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
```

```python
# Parámetros comunes
fs   = 200_000        # Hz (frecuencia de muestreo)
Tc   = 0.05           # s (duración total de la simulación)
t    = np.arange(0, Tc, 1/fs)
Ac   = 1.0            # amplitud de portadora
fc   = 20_000         # Hz (frecuencia de portadora)

mu   = 0.5            # <-- Índice de modulación (ajústalo 0..1 para evitar sobremodulación)

#Señal mensaje tipo coseno
fm_cos = 500          # Hz (frecuencia del coseno del mensaje)

# Señal mensaje tipo pulso rectangular
Tp    = 5e-3          # s (ancho del pulso)
t0    = 10e-3         # s (instante de inicio del pulso)

# Comprobación básica de sobremodulación
mu = float(mu)
if mu < 0:
    print("Aviso: mu<0. Se forzará a 0.")
    mu = 0.0
if mu > 1:
    print("Aviso: mu>1 (sobremodulación en AM-LC). Se forzará a 1 para este ejemplo.")
    mu = 1.0

print(f"Parámetros: fs={fs} Hz, fc={fc} Hz, mu={mu}, fm_cos={fm_cos} Hz, Tp={Tp}s")
```

```python
# Mensaje coseno (rango [-1,1])
m_cos = np.cos(2*np.pi*fm_cos*t)

# Mensaje pulso rectangular (0/1)
m_rect = np.where((t >= t0) & (t < t0+Tp), 1.0, 0.0)

# Opcional: centrado a valor medio cero (para comparar fácilmente en demod)
m_rect_zm = m_rect - np.mean(m_rect)

# Banda aproximada del mensaje (para LPF en demod)
Bm_cos = fm_cos * 1.5      # un poco más del tono
Bm_rect = 2.0 / max(Tp, 1/fs)  # aprox. por transiciones del pulso
```

```python
# AM-LC: s(t) = Ac*(1 + mu*m(t))*cos(2π f_c t)
carrier = np.cos(2*np.pi*fc*t)

s_cos  = Ac*(1 + mu*m_cos )*carrier
s_rect = Ac*(1 + mu*m_rect)*carrier

# Espectros
f_s_cos,  S_cos  = rfft_spectrum(s_cos,  fs)
f_s_rect, S_rect = rfft_spectrum(s_rect, fs)

# Plots (tiempo) — mostrar una ventana corta para ver la portadora
win = (0, 5e-3)  # 5 ms
plot_time(t, [m_cos, s_cos], ['m_cos(t)', 's_cos(t)'],
          xlim=win, title='Tiempo — Mensaje coseno y AM-LC')
plot_time(t, [m_rect, s_rect], ['m_rect(t)', 's_rect(t)'],
          xlim=(t0-2e-3, t0+Tp+2e-3), title='Tiempo — Mensaje rectangular y AM-LC')

# Plots (frecuencia)
plot_freq(f_s_cos,  [S_cos],  ['|S_cos(f)|'],  xlim=(0, 40_000), title='Espectro — AM con coseno')
plot_freq(f_s_rect, [S_rect], ['|S_rect(f)|'], xlim=(0, 40_000), title='Espectro — AM con pulso')
```

```python
# Detección coherente: multiplicar por 2*cos(2π f_c t) y filtrar
y_cos      = 2 * s_cos  * carrier
y_rect     = 2 * s_rect * carrier

# LPF (elige corte según el mensaje)
ycos_lp    = simple_lpf(y_cos,  fs, fc_hz=Bm_cos)
yrect_lp   = simple_lpf(y_rect, fs, fc_hz=Bm_rect)

# Quitar DC para comparar con m(t) (AM-LC deja un término DC ~ Ac)
ycos_base  = ycos_lp  - np.mean(ycos_lp)
yrect_base = yrect_lp - np.mean(yrect_lp)

# Normalización opcional para visual
def norm(x):
    m = np.max(np.abs(x)) + 1e-12
    return x / m

# Plots (tiempo)
plot_time(t, [m_cos, norm(ycos_base)], ['m_cos(t)', 'demod cos (sin DC, norm)'],
          xlim=(0, 15e-3), title='Demodulación coherente — mensaje coseno')
plot_time(t, [m_rect_zm, norm(yrect_base)], ['m_rect(t) (zm)', 'demod rect (sin DC, norm)'],
          xlim=(t0-2e-3, t0+Tp+4e-3), title='Demodulación coherente — mensaje rectangular')

# Espectros de las señales demoduladas (opcional)
f_yc, Myc = rfft_spectrum(ycos_base, fs)
f_yr, Myr = rfft_spectrum(yrect_base, fs)
plot_freq(f_yc, [Myc], ['|Ycos_base(f)|'], xlim=(0, 5_000),
          title='Espectro baseband — coseno demodulado')
plot_freq(f_yr, [Myr], ['|Yrect_base(f)|'], xlim=(0, 5_000),
          title='Espectro baseband — pulso demodulado')
```
