###**1.3Encuentre la **función de densidad espectral** (Transformada de Fourier) para las siguientes señales, sin aplicar propiedades**:

a) $$e^{-a|t|},\quad a \in \mathbb{R}^+$$

b) $$\cos(\omega_c t),\quad \omega_c \in \mathbb{R}$$

c) $$\sin(\omega_s t),\quad \omega_s \in \mathbb{R}$$

d) $$f(t)\cos(\omega_c t),\quad \omega_c \in \mathbb{R},\ f(t)\in \mathbb{R}, \text{ continua}$$

e) $$e^{-a|t|^2},\quad a \in \mathbb{R}^+$$

f) $$A\,\mathrm{rect}\!\left(\frac{t}{d}\right),\quad A,d \in \mathbb{R}$$

Convención y recordatorios

Usaremos la convención (frecuencia angular ω):
$$
X(\omega) = \mathcal{F}\{x(t)\} = \int_{-\infty}^{\infty} x(t)\, e^{-j\omega t}\, dt
$$

$$
x(t) = \frac{1}{2\pi} \int_{-\infty}^{\infty} X(\omega)\, e^{j\omega t}\, d\omega
$$

(a) $$x(t)=e^{-a|t|},\ a>0$$

Partimos de la definición y partimos el eje temporal en dos:

$$
X(\omega)=\int_{-\infty}^{0} e^{at} e^{-j\omega t} dt \;+\; \int_{0}^{\infty} e^{-at} e^{-j\omega t} dt.
$$

Evaluando cada integral (geometría exponencial):

$$
\int_{0}^{\infty} e^{-(a+j\omega)t} dt = \frac{1}{a + j\omega}, \qquad
\int_{-\infty}^{0} e^{(a-j\omega)t} dt = \frac{1}{a - j\omega}.
$$

Sumando:

$$
X(\omega)=\frac{1}{a+j\omega}+\frac{1}{a-j\omega}
=\frac{(a-j\omega)+(a+j\omega)}{a^2+\omega^2}
=\frac{2a}{a^2+\omega^2}.
$$


(b) $$x(t)=\cos(\omega_c t)$$

Definición truncada y límite:

$$
X(\omega)=\lim_{L\to\infty}\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt.
$$

Escribimos $$e^{-j\omega t}=\cos(\omega t)-j\sin(\omega t)$$ y separamos:

$$
\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt
=\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt
 - j\int_{-L}^{L}\cos(\omega_c t)\sin(\omega t)\, dt.
$$

$$\cos(\cdot)\sin(\cdot)$$ es impar en \([ -L, L ] \Rightarrow\) su integral es \(0\).

Queda la parte real:

$$
\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt.
$$

Usamos la identidad trigonométrica (no es propiedad de FT):

$$
\cos A \cos B = \tfrac{1}{2}\big[\cos(A-B)+\cos(A+B)\big].
$$

Entonces

$$
\int_{-L}^{L}\cos(\omega_c t)\cos(\omega t)\, dt
=\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_c-\omega)t\big)\, dt
+\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_c+\omega)t\big)\, dt.
$$

Como $$\int_{-L}^{L}\cos(k t)\, dt = \frac{2\sin(kL)}{k},$$ obtenemos

$$
\int_{-L}^{L}\cos(\omega_c t)\, e^{-j\omega t}\, dt
=\frac{\sin\big((\omega_c-\omega)L\big)}{\omega_c-\omega}
+\frac{\sin\big((\omega_c+\omega)L\big)}{\omega_c+\omega}.
$$

Paso al límite (sentido de distribuciones):

$$
\lim_{L\to\infty}\frac{\sin(L x)}{\pi x} = \delta(x).
$$

Por tanto,

$$
X(\omega)=\pi\big[\delta(\omega-\omega_c)+\delta(\omega+\omega_c)\big].
$$




(c) $$x(t)=\sin(\omega_s t)$$

Definición truncada y límite:

$$
X(\omega)=\lim_{L\to\infty}\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt.
$$

Separando:

$$
\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt
=\int_{-L}^{L}\sin(\omega_s t)\cos(\omega t)\, dt
- j\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt.
$$

$$\sin(\cdot)\cos(\cdot)$$ es impar \(\Rightarrow\) integral \(=0\).

Queda la parte imaginaria:

$$
- j\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt.
$$

Con la identidad

$$
\sin A \sin B = \tfrac{1}{2}\big[\cos(A-B)-\cos(A+B)\big],
$$

tenemos

$$
\int_{-L}^{L}\sin(\omega_s t)\sin(\omega t)\, dt
=\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_s-\omega)t\big)\, dt
-\tfrac{1}{2}\int_{-L}^{L}\cos\big((\omega_s+\omega)t\big)\, dt.
$$

Usando \(\displaystyle\int_{-L}^{L}\cos(k t)\, dt = \frac{2\sin(kL)}{k}\):

$$
\int_{-L}^{L}\sin(\omega_s t)\, e^{-j\omega t}\, dt
=-j\Big[\frac{\sin\big((\omega_s-\omega)L\big)}{\omega_s-\omega}
-\frac{\sin\big((\omega_s+\omega)L\big)}{\omega_s+\omega}\Big].
$$

Paso al límite (distribuciones):

$$
\lim_{L\to\infty}\frac{\sin(L x)}{\pi x} = \delta(x)
\;\Longrightarrow\;

$$
X(\omega) = \frac{\pi}{j}\,\big[\,\delta(\omega-\omega_s)\;-\;\delta(\omega+\omega_s)\,\big].
$$





(d) $$x(t)=f(t)\cos(\omega_c t),\quad f(t)\in\mathbb{R}\ \text{o}\ \mathbb{C}$$

Definimos  
$$F(\Omega)=\int_{-\infty}^{\infty} f(t)\,e^{-j\Omega t}\,dt.$$

Desde la definición:

$$
X(\omega)=\int_{-\infty}^{\infty} f(t)\cos(\omega_c t)\,e^{-j\omega t}\,dt
$$

$$
=\frac{1}{2}\int f(t)\,e^{-j(\omega-\omega_c)t}\,dt
+\frac{1}{2}\int f(t)\,e^{-j(\omega+\omega_c)t}\,dt
$$

$$
=\frac{1}{2}\,F(\omega-\omega_c)+\frac{1}{2}\,F(\omega+\omega_c).
$$

Esto se obtuvo directamente de la definición (simplemente reagrupando el exponente).


(e) $$x(t)=e^{-a\,|t|^{2}}=e^{-a t^{2}},\quad a>0$$

$$
X(\omega)=\int_{-\infty}^{\infty} e^{-a t^{2}}\, e^{-j\omega t}\, dt
= \int e^{-a\left[t^{2}+\frac{j\omega}{a}t\right]}\, dt
$$

$$
= \int \exp\!\left(-a\left(t+\frac{j\omega}{2a}\right)^{2}\right)\,
\exp\!\left(-\frac{\omega^{2}}{4a}\right)\, dt
$$

$$
= \exp\!\left(-\frac{\omega^{2}}{4a}\right)\,
\int_{-\infty}^{\infty} \exp\!\left(-a\left(t+\frac{j\omega}{2a}\right)^{2}\right)\, dt
$$

La integral gaussiana (desplazada) vale $\pi/a$. Por tanto,

$$
X(\omega)=\sqrt{\frac{\pi}{a}}\,\exp\!\left(-\frac{\omega^{2}}{4a}\right).
$$


(f) $$x(t)=A\,\mathrm{rect}_d(t)$$ (pulso de ancho \(d>0\) y amplitud \(A\))

Por definición:

$$
x(t)=
\begin{cases}
A, & |t|<\dfrac{d}{2},\\[6pt]
0, & \text{en otro caso}.
\end{cases}
$$

Entonces

$$
X(\omega)=\int_{-d/2}^{d/2} A\,e^{-j\omega t}\,dt
= A\left[\frac{e^{-j\omega t}}{-j\omega}\right]_{-d/2}^{d/2}
$$

$$
= A\;\frac{e^{-j\omega d/2}-e^{j\omega d/2}}{-j\omega}
= A\;\frac{-2j\sin(\omega d/2)}{-j\omega}
$$

$$
= A\;\frac{2\sin(\omega d/2)}{\omega}
= A\,d\;\mathrm{sinc}\!\left(\frac{\omega d}{2}\right).
$$

En ω=0, por continuidad, $$X(0)=A\,d.$$

###**1.4 Aplique las propiedades de la transformada de Fourier para**
$$\ \omega_1, \omega_c \in \mathbb{R};\ \ B, T \in \mathbb{R}^+.\\[6pt]
\text{Resolver:}
$$


$$
\text{(a)}\quad \mathcal{F}\{u(t)\cos^2(\omega_c t)\}
$$

$$
\text{(b)}\quad \mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}
$$

$$
\text{(c)}\quad \mathcal{F}\{3t^3\}
$$

$$
\text{(d)}\quad \frac{7}{\omega^2 + 6\omega + 45}
$$

$$
\text{(e)}\quad B \displaystyle\sum_{n=-\infty}^{+\infty} \frac{1}{a^2 + (\omega - n\omega_0)^2},
\qquad \text{donde } \omega_0 = \frac{2\pi}{T},\ n \in \{0, \pm1, \pm2, \ldots\}.
$$

\text{Usar Tablas de propiedades y Tablas de Transformada de Fourier.}

###**Propiedades útiles de la Transformada de Fourier**

$$
\text{(1) Modulación:} \quad \mathcal{F}\{e^{-j\omega_1 t} \, x(t)\} = X(\omega + \omega_1)
$$

$$
\text{(2) Desplazamiento en Frecuencia (Inversa):} \quad
\mathcal{F}^{-1}\{X(\omega - \omega_0)\} = e^{j\omega_0 t} x(t)
$$

$$
\text{(3) Producto por } t^n: \quad
\mathcal{F}\{t^n x(t)\} = j^n \frac{d^n}{d\omega^n} X(\omega)
$$

$$
\text{(4) Paso Unitario:} \quad
\mathcal{F}\{u(t)\} = \text{PV}\!\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)
$$

$$
\text{(5) Exponencial Bilateral:} \quad
\mathcal{F}\{e^{-a|t|}\} = \frac{2a}{a^2 + \omega^2}, \quad a>0
$$

$$
\text{(6) Exponencial Unilateral:} \quad
\mathcal{F}\{u(t)e^{-at}\} = \frac{1}{a + j\omega}, \quad a>0
$$

$$
\text{(7) Coseno:} \quad
\mathcal{F}\{\cos(\omega_c t)\} = \pi\left[\delta(\omega - \omega_c) + \delta(\omega + \omega_c)\right]
$$

$$
\text{(8) Periodización en Frecuencia:} \quad
\mathcal{F}^{-1}\left\{\sum_{n=-\infty}^{\infty} X(\omega - n\omega_0)\right\}
=
T \sum_{k=-\infty}^{\infty} x(kT)\,\delta(t - kT), \quad \omega_0 = \frac{2\pi}{T}
$$



**(a)**

$$
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
$$

Usamos la identidad:

$$
\cos^2(\theta)=\frac{1}{2}\left(1+\cos(2\theta)\right)
$$

Entonces:

$$
u(t)\cos^2(\omega_c t)
=
\frac{1}{2}u(t)
+
\frac{1}{2}u(t)\cos(2\omega_c t)
$$

Aplicando transformada y la propiedad de modulación:

$$
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
=
\frac{1}{2}\mathcal{F}\{u(t)\}
+
\frac{1}{2}\cdot\frac{1}{2}\left[U(\omega - 2\omega_c)+U(\omega + 2\omega_c)\right]
$$

donde:

$$
U(\omega)=\mathcal{F}\{u(t)\}
=
\text{PV}\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)
$$

Así, el resultado final es:

$$
\boxed{
\mathcal{F}\{u(t)\cos^2(\omega_c t)\}
=
\frac{1}{2}\left[\text{PV}\left\{\frac{1}{j\omega}\right\} + \pi\delta(\omega)\right]
+
\frac{1}{4}\sum_{s=\pm1}
\left[
\text{PV}\left\{\frac{1}{j(\omega + s\,2\omega_c)}\right\}
+
\pi\delta(\omega + s\,2\omega_c)
\right]
}
$$


**(b)**

$$\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}$$

Por modulación directa:

$$
\mathcal{F}\{\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - \omega_c)+\delta(\omega + \omega_c)\right].
$$

Entonces, usando la propiedad:

$$
\mathcal{F}\{e^{-j\omega_1 t}x(t)\}=X(\omega+\omega_1),
$$

se obtiene:

$$
\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - (\omega_c - \omega_1)) + \delta(\omega + (\omega_c + \omega_1))\right].
$$

$$
\boxed{\mathcal{F}\{e^{-j\omega_1 t}\cos(\omega_c t)\}
=
\pi\left[\delta(\omega - (\omega_c - \omega_1)) + \delta(\omega + (\omega_c + \omega_1))\right]}
$$


**(c)**

$$\mathcal{F}\{3t^3\}$$

Sabemos que:

$$x(t)=1 \quad \Rightarrow \quad X(\omega)=2\pi\delta(\omega).$$

Usando la propiedad del producto por potencias de \(t\):

$$
\mathcal{F}\{t^n x(t)\} = j^n \frac{d^n}{d\omega^n}X(\omega),
$$

entonces, para \(t^3\):

$$
\mathcal{F}\{t^3\}
=
j^3 \frac{d^3}{d\omega^3}\{2\pi \delta(\omega)\}
=
-\,j \, 2\pi\, \delta^{(3)}(\omega).
$$

Multiplicando por 3:

$$
\mathcal{F}\{3t^3\}
=
3\cdot\left(-\,j \, 2\pi\, \delta^{(3)}(\omega)\right)
=
-\,6\pi j \,\delta^{(3)}(\omega).
$$

$$
\boxed{\mathcal{F}\{3t^3\} = -6\pi j\,\delta^{(3)}(\omega)}
$$


**(d)**

$$\mathcal{F}^{-1}\left\{\frac{7}{\omega^2 + 6\omega + 45}\right\}$$

Completamos cuadrado:

$$
\omega^2 + 6\omega + 45 = (\omega+3)^2 + 6^2.
$$

Usamos la propiedad:
$$
\text{Si } X(\omega)=\frac{1}{a^{2}+(\omega-\omega_{0})^{2}}, \quad a>0,
$$

$$
\text{entonces la transformada inversa } x(t)=\frac{1}{2a}\,e^{-a|t|}\,e^{j\omega_{0}t}.
$$


Aquí:

$$a = 6, \qquad \omega_0 = -3.$$

Entonces:

$$
\mathcal{F}^{-1}\left\{\frac{7}{(\omega+3)^2 + 6^2}\right\}
=
7 \cdot \frac{1}{2\cdot 6} \, e^{-6|t|} \, e^{-j3t}
=
\frac{7}{12} e^{-6|t|} e^{-j3t}.
$$

$$
\boxed{\mathcal{F}^{-1}\left\{\frac{7}{\omega^2 + 6\omega + 45}\right\} = \frac{7}{12}\,e^{-6|t|}\,e^{-j3t}}
$$


### (e)

Dado:
$$
F^{-1}\left\{\frac{B}{T}\sum_{n=-\infty}^{\infty}
\left[
\frac{1}{a^{2}+(\omega-n\omega_{0})^{2}}
+
\frac{1}{a+j(\omega-n\omega_{0})}
\right]\right\}
$$

donde $\omega_{0}=\frac{2\pi}{T}.$

#### Base 1:
$$
X_{1}(\omega)=\frac{1}{a^{2}+\omega^{2}}
\quad\Rightarrow\quad
x_{1}(t)=\frac{1}{2a}e^{-a|t|}.
$$

Por periodización en frecuencia:
$$
F^{-1}\left\{\sum_{n=-\infty}^{\infty}X_{1}(\omega-n\omega_{0})\right\}
=
T\sum_{k=-\infty}^{\infty}x_{1}(kT)\,\delta(t-kT)
=
\frac{T}{2a}
\sum_{k=-\infty}^{\infty}e^{-a|k|T}\delta(t-kT).
$$

#### Base 2:
$$
X_{2}(\omega)=\frac{1}{a+j\omega}
\quad\Rightarrow\quad
x_{2}(t)=u(t)e^{-at}.
$$

Periodizando:
$$
F^{-1}\left\{\sum_{n=-\infty}^{\infty}X_{2}(\omega-n\omega_{0})\right\}
=
T\sum_{k=-\infty}^{\infty}x_{2}(kT)\,\delta(t-kT)
=
T\sum_{k=0}^{\infty}e^{-akT}\delta(t-kT).
$$


#### Multiplicando por $ \frac{B}{T} \$:

$$
x(t)=
\frac{B}{T}\left[
\frac{T}{2a}\sum_{k=-\infty}^{\infty}e^{-a|k|T}\delta(t-kT)
+
T\sum_{k=0}^{\infty}e^{-akT}\delta(t-kT)
\right].
$$

$$
\Rightarrow\quad
\boxed{
x(t)=
\frac{B}{2a}\sum_{k=-\infty}^{\infty}e^{-a|k|T}\,\delta(t-kT)
+
B\sum_{k=0}^{\infty}e^{-akT}\,\delta(t-kT)
}
$$
